<h2> Importing The Libraries 


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

<h2> Importing The Data Set 

```python
import os 
os.chdir(r'D:\IIT Tirupati\ML and Data Science Research Basics\2)Regression\Polynomial Regression')
print(os.getcwd())
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values 
y = dataset.iloc[:,-1 ].values
```

<h2> Training The Model on the Linear Regression Model 


```python
from sklearn.linear_model import LinearRegression 
lin_reg = LinearRegression() 
lin_reg.fit(x,y)
```

<h2>Training The Model on the Polynomial Regression Model


```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4) 
x_poly = poly_reg.fit_transform(x)
lin_reg_2= LinearRegression()
lin_reg_2.fit(x_poly , y) 
```

<h2> Visualization Linear Regression Model 


```python
plt.scatter(x,y , color = 'red') 
plt.plot(x , lin_reg.predict(x) , color = 'blue')
plt.title('Truth or Bluff (linear_regression)') 
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show() 
```

<h2> Now Visualizing the Polynomial Regression Graph 

```python
plt.scatter(x,y, color = 'Red')
plt.plot ( x , lin_reg_2.predict(poly_reg.fit_transform(x)) , color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
